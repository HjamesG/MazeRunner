```python
import random
import time

# Need for later

# rooms = each possible section of the map

rooms = [["###   ###",
          "###   ###",
          "###   ###",
          "###   ###",
          "###   ###",
          "###   ###",
          "###   ###",
          "###   ###",
          "###   ###", "UD"],
         
         ["#########",
          "#########",
          "#########",
          "         ",
          "         ",
          "         ",
          "#########",
          "#########",
          "#########", "LR"],
         
         ["###   ###",
          "###   ###",
          "###   ###",
          "###      ",
          "###      ",
          "###      ",
          "#########",
          "#########",
          "#########", "UR"],  
         
         ["###   ###",
          "###   ###",
          "###   ###",
          "      ###",
          "      ###",
          "      ###",
          "#########",
          "#########",
          "#########", "UL"],
         
         ["#########",
          "#########",
          "#########",
          "###      ",
          "###      ",
          "###      ",
          "###   ###",
          "###   ###",
          "###   ###", "RD"],
         
         ["#########",
          "#########",
          "#########",
          "      ###",
          "      ###",
          "      ###",
          "###   ###",
          "###   ###",
          "###   ###", "UD"],
         
         ["###   ###",
          "###   ###",
          "###   ###",
          "         ",
          "         ",
          "         ",
          "###   ###",
          "###   ###",
          "###   ###", "URDL"],
         
         ["###   ###",
          "###   ###",
          "###   ###",
          "###      ",
          "###      ",
          "###      ",
          "###   ###",
          "###   ###",
          "###   ###", "URD"],
         
         ["###   ###",
          "###   ###",
          "###   ###",
          "      ###",
          "      ###",
          "      ###",
          "###   ###",
          "###   ###",
          "###   ###", "UDL"],
         
         ["###   ###",
          "###   ###",
          "###   ###",
          "         ",
          "         ",
          "         ",
          "#########",
          "#########",
          "#########", "URL"],
         
         ["###   ###",
          "###   ###",
          "##     ##",
          "         ",
          "    O    ",
          "         ",
          "##     ##",
          "###   ###",
          "###   ###", "URDL"],
         
         ["#########",
          "#########",
          "#########",
          "         ",
          "         ",
          "         ",
          "###   ###",
          "###   ###",
          "###   ###", "RDL"],
        ]

replaceRoom = ["###   ###",
               "###   ###",
               "##     ##",
               "         ",
               "         ",
               "         ",
               "##     ##",
               "###   ###",
               "###   ###", "URDL"]

endRoom = ["###   ###",
           "###   ###",
           "##     ##",
           "         ",
           "    E    ",
           "         ",
           "##     ##",
           "###   ###",
           "###   ###", "URDL"]

# the player list stores player data like the following:
# world r, world c, direction, screen r, screen c
player = [0, 0, "U", 4, 4]

# represents what time the player starts, initialized here but used later
start = 1

# the playTime represents how much time the player has to escape. 
playTime = 45

# the timer is used to count down from the start time, initialized now used later
timer = 1



# World is a 2d array that stores different rooms for the player to move too. 
world = [[rooms[6]]]
# Representaition of what the world list COULD look like.
# [1, 2, 3],
# [4, 5, 6],
# [7, 8, 9]

# updates the screen with new player information
def printScreenAndPlayer():
    global start
    # creates a copy of the screen that can be modified to be printed for user display
    oldScreen = [world[player[0]][player[1]]]
    newScreen = []
    for i in range(0,9):
        newScreen.append(list(oldScreen[0][i]))
        
    # Shelter time increase    
    if newScreen[player[3]][player[4]] == "O": 
        start = start + 30 
        world[player[0]][player[1]] = replaceRoom
        
    # escape screen
    if newScreen[player[3]][player[4]] == "E": 
        print("\n\n  CONGRATS! \n  YOU ESCAPED THE MAZE!\n\n")
        return "finished"
             
    newScreen[player[3]][player[4]] = "P"
    for r in range(0,9):
        line = ""
        
        for c in range(0,9):
            line = line + newScreen[r][c] + " "
            
        if r == 3:    
            timer = playTime - round(time.time() - start, 1)
            line = line + "       Time Remaining: " + str(timer)
        if r == 4:
            line = line + "       Time is running!"
        print(line)
    newScreen[player[3]][player[4]] = " "

##############################################################################    
# These next 4 methods alter the size of the world 2d list in different ways #
##############################################################################

def expandUp():
    temp = []
    for i in range(0, len(world[0])):
        temp.append("")
    world.insert(0, temp)
    player[0] = player[0] + 1
                    
def expandDown():
    temp = []
    for i in range(0, len(world[0])):
        temp.append("")
    world.append(temp)

def expandRight():
    for i in range(0, len(world)):
        world[i].append("")

def expandLeft():
    for i in range(0, len(world)):
        world[i].insert(0,"")
    player[1] = player[1] + 1
    
# moveUp()    
# checks to see if the player can move up and makes a new room if you cant
    
def moveUp():
    if player[3] == 0:
        if random.randrange(0,5) == 1:
            rooms.append(endRoom)
        if player[0] == 0:
            expandUp()
        player[0] = player[0] - 1
        player[3] = 8
        # make new room that matches direction trying to go
        # plausable list stores rooms that match the direction requirements
        plausable = []
        for x in range(0, len(rooms)):
            for l in range(0,len(rooms[x][-1])):
                if rooms[x][-1][l] == "D":
                    plausable.append(rooms[x])
        world[player[0]][player[1]] = plausable[random.randint(0, len(plausable)-1)]
        
    # makes player unable to move onto the "#"s
    elif world[player[0]][player[1]][player[3]-1][player[4]] == "#":
        return

    else:
    # moves player up
        player[3] = player[3] - 1
     
    
# moves player down if possible, otherwise makes a new room

def moveDown():
    if player[3] == 8:
        if random.randrange(0,5) == 1:
            rooms.append(endRoom)
        if player[0] == len(world)-1:
            expandDown()
        player[0] = player[0] + 1
        player[3] = 0
        # make new room
        # plausable list stores rooms that match the direction requirements
        plausable = []
        for x in range(0, len(rooms)):
            for l in range(0,len(rooms[x][-1])):
                if rooms[x][-1][l] == "U":
                    plausable.append(rooms[x])
        world[player[0]][player[1]] = plausable[random.randint(0, len(plausable)-1)]
        
    # makes player unable to move onto the "#"s
    elif world[player[0]][player[1]][player[3]+1][player[4]] == "#":
        return
        
    else:
    # moves player down
        player[3] = player[3] + 1

# Checks if player can move left otherwise makes new room

def moveLeft():
    if player[4] == 0:
        if random.randrange(0,5) == 1:
            rooms.append(endRoom)
        if player[1] == 0:
            expandLeft()
        player[1] = player[1] - 1
        player[4] = 8
        # make new room
        # plausable list stores rooms that match the direction requirements
        plausable = []
        for x in range(0, len(rooms)):
            for l in range(0,len(rooms[x][-1])):
                if rooms[x][-1][l] == "R":
                    plausable.append(rooms[x])
        world[player[0]][player[1]] = plausable[random.randint(0, len(plausable)-1)]
        
    # makes player unable to move onto the "#"s
    elif world[player[0]][player[1]][player[3]][player[4]-1] == "#":
        return

    else:
    # moves player left
        player[4] = player[4] - 1
    
# checks to see if player can move right and makes a room if it cannot

def moveRight():
    if player[4] == 8:
        if random.randrange(0,5) == 1:
            rooms.append(endRoom)
        if player[1] == len(world[0])-1:
            expandRight()
        #if world[player[0]][player[1]+1] == "":
        player[1] = player[1] + 1
        player[4] = 0
        # make new room
        # plausable list stores rooms that match the direction requirements
        plausable = []
        for x in range(0, len(rooms)):
            for l in range(0,len(rooms[x][-1])):
                if rooms[x][-1][l] == "L":
                    plausable.append(rooms[x])
        world[player[0]][player[1]] = plausable[random.randint(0, len(plausable)-1)]
        
    # makes player unable to move onto the "#"s
    elif world[player[0]][player[1]][player[3]][player[4]+1] == "#":
        return

    else:
    # moves player right
        player[4] = player[4] + 1
    
# game loop triggered when game started

def gamePlay():    
    while True:
        if playTime <= (time.time() - start):
            print("\n\n GAME OVER. YOU DIED.\n\n")
            return
        if printScreenAndPlayer() == "finished":
            break
        usr = input("")
        if usr == "w":
            moveUp()
        elif usr == "s":
            moveDown()
        elif usr == "a":
            moveLeft()
        elif usr == "d":
            moveRight()

            
# Main menu loop

while True:
    print("\nMAIN MENU:")
    print("[1] Start Game")
    print("[2] Exit")
    print("[3] Instructions")
    usr = int(input("What would you like to do? "))
    if usr == 1:
        start = int(time.time())
        playTime = 45
        gamePlay()
    elif usr == 2:
        print("Goodbye!")
    elif usr == 3:
        print("\nYou are the player (P) and you are trying to get to the")
        print("Exit (E) before your time runs out. Your time will increase")
        print("if you find a shelter (O). Keep your time above 0 and find")
        print("the exit to win! \n\n-H.G. + A.W.\n")
        
```
